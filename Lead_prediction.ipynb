```python
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv('/content/Unified_Leads.csv')
print(a)
print(a.shape)
print(a.size)
print("#############################")
print(a.isna().sum())
print(a.head())
print(a.tail())
print(a.dtypes)
```

```python
d=a.drop(['Name','contact_number','Contact_email','lead_tag','references','creation_date','first_call_attempt','last_call_date','last_call_time'],axis=1)
d
```

```python
d.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
d['lead_stage']=le.fit_transform(d['lead_stage'])
d['assigned_to']=le.fit_transform(d['assigned_to'])
d['campaign_name']=le.fit_transform(d['campaign_name'])
d['source_type']=le.fit_transform(d['source_type'])
d['source_name']=le.fit_transform(d['source_name'])
d
```

```python
d['lead_status']=d['lead_status'].apply(lambda x: 1 if x == 'CLOSED' else 0)
d
```

```python
x=d.iloc[:,1:].values
x
```

```python
y=d.iloc[:,:1].values
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30)
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
s.fit(xtest)
xtrain=s.transform(xtrain)
xtest=s.transform(xtest)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
svm_model=SVC()
nb_model=MultinomialNB()
knn_model=KNeighborsClassifier(n_neighbors=7)
rf_model=RandomForestClassifier(n_estimators=100)
dt_model=DecisionTreeClassifier()
lstmodel=[svm_model,nb_model,knn_model,rf_model,dt_model]
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_auc_score
for i in lstmodel:
  print(i)
  i.fit(xtrain,ytrain)
  ypred=i.predict(xtest)
  ac=accuracy_score(ypred,ytest)
  roc=roc_auc_score(ypred,ytest)
  print("######################")
  print(classification_report(ypred,ytest))
  print("accuracy score",ac)
  print("roc auc score",roc)
  print("******************************")
```
